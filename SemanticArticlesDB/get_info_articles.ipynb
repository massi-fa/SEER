```python
import json
import pandas as pd

# Carica il file JSON
with open('news_semantic_db.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# Estrai la lista degli articoli
articles = data['articles']

# Crea un DataFrame pandas
df = pd.DataFrame(articles)

# Salva in Excel
df.to_excel('articles_info.xlsx', index=False)

print("Excel creato con successo: articles_info.xlsx")
print(f"Numero di articoli: {len(df)}")
print("Colonne:", list(df.columns))
```
